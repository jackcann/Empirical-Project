```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import time
import random
```

```python
urls={'Premier League': 'https://fbref.com/en/comps/9/history/Premier-League-Seasons',
    'La Liga': 'https://fbref.com/en/comps/12/history/La-Liga-Seasons',
    'Serie A': 'https://fbref.com/en/comps/11/history/Serie-A-Seasons',
    'Bundesliga': 'https://fbref.com/en/comps/20/history/Bundesliga-Seasons',
    'Ligue 1': 'https://fbref.com/en/comps/13/history/Ligue-1-Seasons'}

#This function gets the urls for each season from the history page
def get_urls(history_url, season_number=15):
    page=requests.get(history_url)
    soup=BeautifulSoup(page.content,'html.parser')
    table=soup.find('table')
    urls=[]
    
    for row in table.find('tbody').find_all('tr'):
        first_cell=row.find('th')
        if first_cell and first_cell.find('a'):
            link=first_cell.find('a')['href']
            full_link='https://fbref.com'+link
            urls.append(full_link)
    
    return urls[:season_number]

#This list stores all the data retrieved
all_data=[]

#Starts from most recent season in order to later count down
starting_year=2025

for league, history_url in urls.items():
    print(f"\n webscraping {league}")
    
    urls=get_urls(history_url)
    year=starting_year
    
    for season_url in urls:
        try:
            tables=pd.read_html(season_url)
            league_table=tables[0]
            time.sleep(random.randint(4, 8))

            #Organise table
            league_table2=league_table[['Rk', 'Squad', 'Pts', 'MP', 'Pts/MP',
                                        'W', 'D', 'L']]
            league_table2.columns=['Position', 'Team', 'Points', 
                                    'Matches Played', 'Points per Match',
                                   'Wins', 'Draws', 'Losses']
            league_table2=league_table2.dropna()

            #Create league table and season year
            league_table2['League']=league
            league_table2['Season']=year

            #Format columns
            league_table2=league_table2[['League', 'Season', 'Position',
                                         'Team', 'Points', 'Matches Played', 
                                         'Points per Match', 'Wins', 'Draws', 'Losses']]

            all_data.append(league_table2)

            print(f"Scraped {league} {year}")
            #Count down the seasons
            year-=1 
        
        except Exception as e:
            print(f"Error in {league} {year}: {e}")


#Put into one dataframe
combined_df=pd.concat(all_data, ignore_index=True)

#Create CSV
combined_df.to_csv('data.csv', index=False)

print("\n Data saved")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Function to calculate entropy
def calculate_entropy(df_season):
    # Normalize points per match to get probabilities
    total_points = df_season['Points per Match'].sum()
    df_season['p_i'] = df_season['Points per Match'] / total_points
    
    # Calculate entropy
    entropy = -(df_season['p_i'] * np.log2(df_season['p_i'])).sum() / np.log2(len(df_season))
    return entropy

# Function to calculate penalty term
def calculate_penalty_term(df_season):
    # Calculate mean and standard deviation of points per match
    mean_points = df_season['Points per Match'].mean()
    std_points = df_season['Points per Match'].std()
    
    # Calculate penalty term
    penalty_term = max(0, 1 - (std_points / mean_points))
    return penalty_term

# Function to calculate HHI
def calculate_hhi(df_season):
    # Normalize points per match to get probabilities
    total_points = df_season['Points per Match'].sum()
    df_season['p_i'] = df_season['Points per Match'] / total_points
    
    # Calculate HHI
    hhi = 1 - (df_season['p_i'] ** 2).sum()
    return hhi

# Function to calculate competitiveness score
def calculate_competitiveness(df_season):
    # Calculate the components
    entropy = calculate_entropy(df_season)
    penalty_term = calculate_penalty_term(df_season)
    hhi = calculate_hhi(df_season)
    
    # Weighted sum to get competitiveness score
    competitiveness_score = (0.4 * entropy) + (0.3 * penalty_term) + (0.3 * hhi)
    
    return competitiveness_score

# Empty list to store the competitiveness scores for each season
competitiveness_scores = []

# Loop through each league and season in the dataframe
for league in df['League'].unique():
    for season in df[df['League'] == league]['Season'].unique():
        df_season = df[(df['League'] == league) & (df['Season'] == season)]
        
        # Calculate competitiveness score for this season
        score = calculate_competitiveness(df_season)
        
        # Append the result to the list
        competitiveness_scores.append({
            'League': league,
            'Season': season,
            'Competitiveness Score': score
        })

# Create a DataFrame with the competitiveness scores
competitiveness_df = pd.DataFrame(competitiveness_scores)

# View the result
print(competitiveness_df)

# Plotting the competitiveness scores
plt.figure(figsize=(10, 6))
sns.lineplot(data=competitiveness_df, x='Season', y='Competitiveness Score', hue='League')
plt.title('Competitiveness Score Over Time for Each League')
plt.xlabel('Season')
plt.ylabel('Competitiveness Score')
plt.show()
```
